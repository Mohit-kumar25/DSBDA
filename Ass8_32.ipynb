```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# make plots appear in notebook
%matplotlib inline
```

```python
titanic = pd.read_csv("/home/csl-4/Downloads/train.csv")
```

```python
titanic.head()
```

```python
titanic.shape
```

```python
titanic.info()
```

```python
sns.countplot(x='Survived', data=titanic)
plt.title("Survival Count")
plt.show()
```

```python
sns.countplot(x='Survived', hue='Sex', data=titanic)
plt.title("Survival by Gender")
plt.show()
```

```python
sns.countplot(x='Survived', hue='Pclass', data=titanic)
plt.title("Survival by Passenger Class")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(titanic['Age'].dropna(), bins=30, kde=True)
plt.title("Age Distribution of Passengers")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(titanic.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.histplot(titanic['Fare'], bins=30, kde=True)

plt.title("Distribution of Ticket Prices")
plt.xlabel("Fare")
plt.ylabel("Number of Passengers")

plt.show()
```
