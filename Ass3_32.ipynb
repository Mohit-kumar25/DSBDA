```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/home/csl-4/Downloads/train.csv")
df
```

```python
df.info()
```

```python
categorical_var = "Pclass"   # ticket class
numeric_vars = ["Age", "Fare"]
```

```python
df[numeric_vars] = df[numeric_vars].fillna(df[numeric_vars].median())
```

```python
df[categorical_var] = df[categorical_var].fillna(df[categorical_var].mode()[0])
```

```python
df[numeric_vars].describe()
```

```python
grouped_stats = df.groupby(categorical_var)[numeric_vars].agg(
    ["mean", "median", "min", "max", "std"]
)
grouped_stats
```

```python
grouped_stats.columns = [
    f"{col}_{stat}" for col, stat in grouped_stats.columns
]

grouped_stats
```

```python
for var in numeric_vars:
    print(f"\nSummary of {var} grouped by {categorical_var}:")
    print(df.groupby(categorical_var)[var].describe())
```

```python
age_list_by_class = df.groupby(categorical_var)["Age"].apply(list)
fare_list_by_class = df.groupby(categorical_var)["Fare"].apply(list)

print("Age lists by class:\n", age_list_by_class)
print("\nFare lists by class:\n", fare_list_by_class)
```

```python
# Load Iris dataset 
data = pd.read_csv('/home/csl-4/Downloads/Iris.csv')  

# Display 
data.head()
```

```python
# Check the column names
print(data.columns)

# Clean any extra spaces in column names if needed
data.columns = data.columns.str.strip()

print(data.columns)
```

```python
# Filter
setosa = data[data['Species'] == 'Iris-setosa']
versicolor = data[data['Species'] == 'Iris-versicolor']
virginica = data[data['Species'] == 'Iris-virginica']
```

```python
# Function to calculate and display statistics (mean, std, percentiles) 
def display_statistics(species_data, species_name):
    print(f"\nStatistics for {species_name}:")
    print(f"Mean:")
    print(species_data.mean())
    print(f"\nStandard Deviation:")
    print(species_data.std())
    print(f"\nPercentiles:")
    print(f"25th Percentile: {np.percentile(species_data, 25)}")
    print(f"50th Percentile (Median): {np.percentile(species_data, 50)}")
    print(f"75th Percentile: {np.percentile(species_data, 75)}")
```

```python
# Displaying for Iris-setosa
display_statistics(setosa.drop(columns=['Id', 'Species']), 'Iris-setosa')

# Displaying for Iris-versicolor
display_statistics(versicolor.drop(columns=['Id', 'Species']), 'Iris-versicolor')

# Displaying for Iris-virginica
display_statistics(virginica.drop(columns=['Id', 'Species']), 'Iris-virginica')
```
