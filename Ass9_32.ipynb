```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# make plots appear in notebook
%matplotlib inline
```

```python
titanic = pd.read_csv("/home/csl-4/Downloads/train.csv")
titanic.head()
```

```python
plt.figure(figsize=(8,6))

sns.boxplot(x='Sex', y='Age', hue='Survived', data=titanic)

plt.title("Age Distribution by Gender and Survival")
plt.xlabel("Gender")
plt.ylabel("Age")

plt.show()
```

```python
# Average age of passengers by gender and survival
stats = titanic.groupby(['Sex','Survived'])['Age'].mean()

print("Average Age by Gender and Survival:\n")
print(stats)
```

```python
survival_counts = titanic.groupby('Sex')['Survived'].value_counts()

print("Survival Count by Gender:\n")
print(survival_counts)
```

```python
print("Observations from the Analysis:\n")

print("1. Female passengers have a much higher survival rate compared to male passengers.")

print("2. A large number of male passengers did not survive the disaster.")

print("3. Younger passengers tend to appear more among survivors.")

print("4. The average age of male passengers who died is higher compared to male survivors.")

print("5. The boxplot and statistics together indicate that gender played a major role in survival probability.")
```
